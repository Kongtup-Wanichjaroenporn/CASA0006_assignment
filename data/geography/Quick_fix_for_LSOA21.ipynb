```python
from pathlib import Path
import geopandas as gpd

DATA = Path('large_data')
LSOA_GPKG = DATA / 'LSOA21_EW_BFC_V10.gpkg' #from this link https://geoportal.statistics.gov.uk/datasets/ons::lower-layer-super-output-areas-december-2021-boundaries-ew-bfc-v10-2/about

LONDON_BOROUGHS = {
    'City of London', 'Barking and Dagenham', 'Barnet', 'Bexley', 'Brent',
    'Bromley', 'Camden', 'Croydon', 'Ealing', 'Enfield', 'Greenwich',
    'Hackney', 'Hammersmith and Fulham', 'Haringey', 'Harrow', 'Havering',
    'Hillingdon', 'Hounslow', 'Islington', 'Kensington and Chelsea',
    'Kingston upon Thames', 'Lambeth', 'Lewisham', 'Merton', 'Newham',
    'Redbridge', 'Richmond upon Thames', 'Southwark', 'Sutton',
    'Tower Hamlets', 'Waltham Forest', 'Wandsworth', 'Westminster',
}

lsoa = gpd.read_file(LSOA_GPKG)

# Borough name = LSOA21NM with the trailing " NNNX" stripped (e.g. "Camden 001A" -> "Camden")
lsoa['borough'] = lsoa['LSOA21NM'].str.replace(r'\s\d{3}[A-Z]$', '', regex=True)

london_lsoa_gdf = lsoa[lsoa['borough'].isin(LONDON_BOROUGHS)].copy()

print(f'Filtered: {len(london_lsoa_gdf):,} LSOAs across {london_lsoa_gdf["borough"].nunique()} boroughs')
print(f'CRS: {london_lsoa_gdf.crs}')
london_lsoa_gdf[['LSOA21CD', 'LSOA21NM', 'borough']].head()
```

```python
out = 'LSOA21_London.gpkg'
london_lsoa_gdf.to_file(out, driver='GPKG')
```
